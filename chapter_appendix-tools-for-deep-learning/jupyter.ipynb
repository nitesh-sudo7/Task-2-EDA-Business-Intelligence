```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.show()
```

```python
## Dataset Loading
#The cleaned dataset obtained from Task 1 is loaded for further analysis.
df = pd.read_csv("/content/cleaned_data.csv")
df.head()
```

```python
## Dataset Overview
#This section provides a high-level understanding of the dataset including shape, column names, and data types.

df.shape
df.columns
df.info()
```

```python
## Descriptive Statistics
#Summary statistics help understand central tendency, spread, and distribution of numerical features.
df.describe()
df.select_dtypes(include='object').describe()
```

```python
## Univariate Analysis

#Univariate analysis examines individual variables to understand their distributions and frequency patterns.
df = pd.read_csv("/content/cleaned_data.csv")
plt.figure(figsize=(6,4))
sns.histplot(df['Total_Amount'], kde=True)
plt.title("Revenue Distribution")
plt.show()

#Insight: Revenue appears to be right-skewed, indicating a small number of high-value transactions.
```

```python
## Multivariate Analysis

#This section explores relationships between multiple variables to identify correlations and trends.

plt.figure(figsize=(6,4))
sns.scatterplot(x='Quantity', y='Total_Amount', data=df)
plt.title("Quantity vs Revenue")
plt.show()
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()

#Insight: Quantity and revenue show a strong positive correlation, suggesting higher sales volumes directly impact revenue.
```

## Key Insights from EDA

- Revenue shows noticeable variability with some high-value outliers.
- A limited number of products contribute to a significant portion of total revenue.
- Strong correlation observed between quantity sold and revenue.
- Certain categories/regions consistently outperform others.
